## **UNIVARIATE ANALYSIS**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')
```

```python
# Assuming 'df' is your DataFrame
WATT_counts = df['WATT'].value_counts()

# Using Matplotlib to create a count plot
plt.figure(figsize=(8, 6))
plt.bar(WATT_counts.index, WATT_counts, color='green')
plt.title('Count WATT QUALITY')
plt.xlabel('WATT')
plt.ylabel('Count')
plt.show()
```

 From the plot above,we notice that 600W was sold more than any other Watt.

## **USING SWARM PLOT TO UNDERSTAND THE RELATIONSHIP BETWEEN THE QUANTITY AND WATT**

```python
# Assuming 'df' is your DataFrame
plt.figure(figsize=(10, 8))

# Using Seaborn to create a swarm plot
sns.swarmplot(x="WATT", y="QTY", data=df, palette='viridis')

plt.title('Swarm Plot for WATT and QTY')
plt.xlabel('WATT')
plt.ylabel('QTY')
plt.show()
```

According to the swarm plot, even though 600 WATT was sold more, 455Watt had the highest quantity sold.

## **USING BOXPLOT TO VISUALIZE DATA**

```python
# Check the column names of the DataFrame
print(df.columns)

# Load the data again with the correct column names
df = pd.read_csv("/content/SOLAR PANEL SALES 2023.csv", names=["DATE", "WATT", "QTY", "RATE"])
```

```python
# Create boxplots for WATT and QTY
fig, axs = plt.subplots(2, 2, figsize=(15, 8))
sns.boxplot(df['WATT'], ax=axs[0, 0])
sns.boxplot(df['QTY'], ax=axs[0, 1])
plt.show()
```

 According to the boxplot,the WATT has no outliers while the Quantity has normal distribution with few outliers.

## **USING HISTOGRAM TO CHECK DISTRIBUTION**

```python
# Create histograms for WATT and QTY
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
sns.histplot(data=df, x="WATT", kde=True, color="skyblue", ax=axs[0])
sns.histplot(data=df, x="QTY", kde=True, color="olive", ax=axs[1])
plt.show()
```

This analysis was useful in analyzing the sales volumn of each watt and their growth which will help improve sales for 2024
