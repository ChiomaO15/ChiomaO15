## **IMPORT DATASET WHICH CONTAINS GENDER,  HEIGHT, WEIGHT**

```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("/content/bmi.csv")
```

```python
print(df)
```

```python
df['Gender'] = df['Gender'].astype('category')
print(df)
print(df.dtypes)
```

## **CHANGING STRINGS TO FLOAT**

```python
df['Gender'] = df['Gender'].replace('Male', 0)
df['Gender'] = df['Gender'].replace('Female', 1)
df['Gender'] = df['Gender'].astype(float)
print(df)
```

```python
df['Gender'] = df['Gender'].astype('category')
print(df)
```

## **SEPARATING X_TRAIN AND Y_TRAIN**

```python
y = df['Index']
y
```

```python
X = df.drop('Index',axis =1)
X
```

## **DATA PREPARATION**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = 0.26)
```

```python
X_train
```

```python
X_test
```

#**MODEL BUILDING**

# **LINEAR REGRESSION**

 **TRAINING THE MODEL**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred  = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

## **EVALUATE MODEL PERFORMANCE**

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results = pd.DataFrame(['Linear regression',lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2','TEST_MSE','TEST R2']
```

```python
lr_results
```

# **RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth =2, random_state=100)
rf.fit(X_train, y_train)
```

###**EVALUATION OF MODEL**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred  = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest',rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2','TEST_MSE','TEST R2']
rf_results
```

# **COMPARISON OF LOGISTIC REGRESSION AND RANDOM FOREST**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```
